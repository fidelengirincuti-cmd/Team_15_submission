# Technical Architecture
# Feature Engineering
The system uses a dual-feature extraction approach optimized for multilingual text:

- Word-level TF-IDF (1-2 grams): Captures semantic patterns across languages
- Character-level TF-IDF (2-5 grams): Captures Kinyarwanda morphological patterns

# Model Performance

- Best Model: K-Nearest Neighbors (KNN) with cosine distance
- Test Accuracy: 76.51%
- Training Samples: 126,817 business descriptions
- Classification Categories: 396 unique ISIC codes
C- lass Balancing: Weighted to handle imbalanced code distributions

#  Requirements
Training Data Format

-File: CSV with ISO-8859-1 encoding
- Required Columns:
    - D03B: Business description text (multilingual)
    - D03B1: ISIC code (target variable)

-Minimum: 2+ samples per ISIC code for training

Prediction Data Format

- Single Prediction: Plain text string
- Batch Prediction: CSV with at least one text column

Model Confidence Interpretation

- High Confidence (>70%): Prediction suitable for automatic coding
- Moderate Confidence (50-70%): Review recommended
- Low Confidence (<50%): Manual verification required

# Technical Notes
Performance Considerations

- Memory Usage: ~500MB for loaded model

- Batch Processing: ~1 second per 100 rows

Known Limitations

- Character encoding must be UTF-8 or ISO-8859-1
- Maximum input text length: 10,000 characters
- Rare ISIC codes (<5 training samples) may have lower accuracy
